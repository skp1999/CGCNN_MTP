```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_binary = pd.read_excel ('image_new.xlsx', sheet_name='Binary_image',header=None)
df_binary
```

```python
x = []
for i in range(206):
    img = df_binary[i]
    img = np.array(img)
    img = img.reshape((128,128))
    x.append(img)

x = np.array(x)
```

```python
x.shape
```

```python
X_train = x[0:180]
X_test = x[180:]
```

```python
X_train = X_train.reshape(180,128,128,1)
X_test = X_test.reshape(26,128,128,1)

print(X_train.shape, X_test.shape)
```

```python
properties = pd.read_excel('properties_new.xlsx')
porosity = properties['Porosity']
y = properties['Diffusivity']
y = y/(1.958*10**(-14))
```

```python
y = np.array(y)
y_train = y[0:180]
y_test = y[180:]
print(y_train.shape, y_test.shape)
```

```python
y_train = y_train.reshape(180,1)
y_test = y_test.reshape(26,1)
```

```python
print(y_train.shape, y_test.shape)
```

### Model -1 (relu, no padding, lr=-3)

```python
model = keras.models.Sequential()


model.add(Conv2D(filters=6, kernel_size=(5,5), activation='relu',input_shape=(128,128,1)))
model.add(MaxPool2D(pool_size=(4,4)))
model.add(Conv2D(filters=12, kernel_size=(1,1), activation='relu'))
model.add(MaxPool2D(pool_size=(4,4)))

model.add(Flatten())

model.add(Dense(128))
model.add(Dropout(0.25))
# model.add(Activation('relu'))


model.add(Dense(1))
model.add(Dropout(0.5))
```

```python
tf.keras.utils.plot_model(
    model,
    to_file="model.png",
    show_shapes=False,
    show_dtype=False,
    show_layer_names=True,
    rankdir="TB",
    expand_nested=False,
    dpi=96
)
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mape'])
checkpoint_filepath = 'checkpoint_new/model_1'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=True,
    monitor='mape',
    mode='min',
    save_best_only=True)
#model.compile(optimizer='adam', loss='mse', metrics=['mape'])
history = model.fit(X_train,y_train, epochs=50, steps_per_epoch=len(X_train), validation_steps=len(X_test))
```

```python
model.save('model_new/model_1.h5')
```

```python
new_model = tf.keras.models.load_model('model_new/model_1.h5')

# Show the model architecture
new_model.summary()
```

```python
train_mape = new_model.evaluate(X_train, y_train, verbose=2)[1]
test_mape = new_model.evaluate(X_test, y_test, verbose=2)[1]
print('train mape : {:5.2f}'.format(train_mape))
print('test mape : {:5.2f}'.format(test_mape))
```

### Model - 8 

```python
model = keras.models.Sequential()


model.add(Conv2D(filters=6, kernel_size=(3,3), activation='relu', input_shape=(128,128,1)))
model.add(MaxPool2D(pool_size=(4,4)))
model.add(Conv2D(filters=12, kernel_size=(1,1), activation='relu'))
model.add(MaxPool2D(pool_size=(4,4)))

model.add(Flatten())

model.add(Dense(256))
model.add(Dropout(0.1))
model.add(Activation('relu'))

model.add(Dense(128))
model.add(Dropout(0.25))
model.add(Activation('relu'))


model.add(Dense(1))
model.add(Dropout(0.5))
```

```python
model.summary()
```
