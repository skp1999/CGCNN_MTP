```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity
```

```python
img_path = 'Grayscale_images/img_0.jpg'
img_gray = cv2.imread(img_path,0)
plt.imshow(img_gray, cmap='gray')
plt.show()

img_bw = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)[1]
plt.imshow(img_bw, cmap='gray')
plt.show()

img_path = 'Binary_images/img_0.jpg'
img_bin = cv2.imread(img_path,0)
plt.imshow(img_bin, cmap='gray')
plt.show()

(score, diff) = structural_similarity(img_bin, img_bw, full=True)
print("Image similarity", score)
```

```python
similarity_scores = []
for i in range(200):
    img_gray = cv2.imread('Grayscale_images/img_'+str(i)+'.jpg',0)
    img_bw = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)[1]
    img_bin = cv2.imread('Binary_images/img_'+str(i)+'.jpg',0)
    similarity_scores.append(structural_similarity(img_bin, img_bw, full=True)[0])
```

```python
scores = np.array(similarity_scores)
```

```python
np.min(scores)
```

```python
np.max(scores)
```

```python
x = [i for i in range(200)]
y = similarity_scores
```

```python
plt.plot(x,y)
```

```python
from scipy.ndimage.filters import gaussian_filter1d

ysmoothed = gaussian_filter1d(y, sigma=2)
plt.xlabel('Image  number')
plt.ylabel('Similarity')
plt.plot(x, ysmoothed)
plt.show()
```
