### Model -1 (relu, no padding, lr=-3)

```python
model = keras.models.Sequential()


model.add(Conv2D(filters=6, kernel_size=(5,5), activation='relu',input_shape=(128,128,1)))
model.add(MaxPool2D(pool_size=(4,4)))
model.add(Conv2D(filters=12, kernel_size=(1,1), activation='relu'))
model.add(MaxPool2D(pool_size=(4,4)))

model.add(Flatten())

model.add(Dense(128))
model.add(Dropout(0.25))
# model.add(Activation('relu'))


model.add(Dense(1))
model.add(Dropout(0.5))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mape'])

checkpoint_filepath = 'checkpoint_old/model_1'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=True,
    monitor='mape',
    mode='min',
    save_best_only=True)

history = model.fit(X_train,y_train, epochs=50, callbacks=[model_checkpoint_callback], steps_per_epoch=len(X_train), validation_steps=len(X_test))
```

```python
model.load_weights(checkpoint_filepath)
```

```python
model.save('model_old/model_1.h5')
```

```python
new_model = tf.keras.models.load_model('model_old/model_1.h5')

# Show the model architecture
new_model.summary()
```

```python
train_mape = new_model.evaluate(X_train, y_train, verbose=2)[1]
test_mape = new_model.evaluate(X_test, y_test, verbose=2)[1]
print('train mape : {:5.2f}'.format(train_mape))
print('test mape : {:5.2f}'.format(test_mape))
```
