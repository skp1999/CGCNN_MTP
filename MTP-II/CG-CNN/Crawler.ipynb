```python
import requests
import lxml
from bs4 import BeautifulSoup
```

```python
url = 'https://materialsproject.org/materials/mp-26639/'
f = requests.get(url)
```

```python
soup = BeautifulSoup(f.content,'lxml')
```

```python
content = soup.find('table',{'class':'elastic-table table table-bordered'}).find_all()
```

```python
tc = content[0].find_all('td')
for i in range(len(tc)):
    if((tc[i].find('h5').text).strip() == 'Bulk Modulus KVRH'):
        bulk = int((tc[i].find('span').text).split()[0])
    if((tc[i].find('h5').text).strip() == 'Shear Modulus GVRH'):
        shear = int((tc[i].find('span').text).split()[0])
print(bulk, shear)   
```
