```python
# imported the requests library
import requests
image_url = "https://materialsproject.org/materials/mp-1020797/cif"
  
# URL of the image to be downloaded is defined as image_url
r = requests.get(image_url) # create HTTP response object

  
# send a HTTP request to the server and save
# the HTTP response in a response object called r
with open("1.cif",'wb') as f:
  
    # Saving received content as a png file in
    # binary format
  
    # write the contents of the response (r.content)
    # to a new file in binary mode.
    f.write((r.content))
```

```python
import numpy as np
import pandas as pd
import requests

data = pd.read_csv('predict.csv')
data = list(np.array(data['Material_ID']))
i=0

for material in data:
    print(i, material)
    i+=1
    link = "https://materialsproject.org/materials/" + material + "/cif"
    r = requests.get(link)
    filename = 'materials_cif_test/'+material+'.cif'
    with open(filename,'wb') as f:
        f.write((r.content))
```
