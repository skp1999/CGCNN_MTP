```python
import json
file = "materials_data_full/materials_data.json"
with open(file) as train_file:
    dict_train = json.load(train_file)

# # converting json dataset from dictionary to dataframe
# train = pd.DataFrame.from_dict(dict_train, orient='index')
# train.reset_index(level=0, inplace=True)
```

```python
train = pd.DataFrame.from_dict(dict_train['response'], orient='index')
train.reset_index(level=0, inplace=True)
train
```

```python
data = {}
for key in dict_train.keys():
    if(key=='mp-23013'):
        continue
    data[key] = {}
    if(dict_train[key]['response']==[]):
        data[key]['Bulk Modulus'] = ""
        data[key]['Shear Modulus'] = ""
        data[key]['Comment'] = 'Material Not found'
    else:
        if(dict_train[key]['response'][0]['elasticity']==None):
            data[key]['Bulk Modulus'] = ""
            data[key]['Shear Modulus'] = ""
            data[key]['Comment'] = 'Elasticity not calculated'
        else:
            data[key]['Bulk Modulus'] = dict_train[key]['response'][0]['elasticity']['K_VRH']
            data[key]['Shear Modulus'] = dict_train[key]['response'][0]['elasticity']['G_VRH']
            data[key]['Comment'] = 'Elasticity values obtained'
```

```python
df = pd.DataFrame(columns = ['Material_ID','Bulk Modulus','Shear Modulus','Comment'])
i=0
for key in data.keys():
    print(i)
    i+=1
    df2 = {'Material_ID':key, 'Bulk Modulus': data[key]['Bulk Modulus'], 
           'Shear Modulus': data[key]['Shear Modulus'], 'Comment': data[key]['Comment']}
    df = df.append(df2, ignore_index = True)
```

```python
df
```

```python
df.to_csv('materials_bulk_shear.csv',index=False)
```
